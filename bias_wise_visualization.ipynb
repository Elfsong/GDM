```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
domain = "Age"
# Generate random data for demonstration
data1 = np.random.normal(0.22, 0.01, 1000) + np.random.normal(0.01, 0.005, 1000)
data2 = np.random.normal(0.26, 0.05, 1000) + np.random.normal(0.01, 0.005, 1000)
data3 = np.random.normal(0.13, 0.04, 1000) + np.random.normal(0.01, 0.005, 1000) 
data4 = np.random.normal(0.06, 0.03, 1000) + np.random.normal(0.01, 0.005, 1000)
data5 = np.random.normal(0.07, 0.02, 1000) + np.random.normal(0.01, 0.005, 1000) 

bins = np.arange(-0.05, 0.4, 0.02)

plt.figure(figsize=(5, 3))

# Create the histogram using Seaborn
sns.histplot(data1, bins=bins, kde=False, edgecolor='white', color='#d00000', alpha=0.6, label='Baseline')
sns.histplot(data2, bins=bins, kde=False, edgecolor='white', color='#dcf763', alpha=0.6, label='Sequential')
sns.histplot(data3, bins=bins, kde=False, edgecolor='white', color='#a2aebb', alpha=0.6, label='Voting')
sns.histplot(data4, bins=bins, kde=False, edgecolor='white', color='#ffba08', alpha=0.6, label='Debating')
sns.histplot(data5, bins=bins, kde=False, edgecolor='white', color='#3f88c5', alpha=0.6, label='Committee')

# Add a title and labels
plt.title(domain)
plt.xlabel("")
plt.ylabel("")

# Add a vertical line at x=0.1
plt.axvline(x=0.22, color='#d00000', linestyle='--', linewidth=2)
# plt.axvline(x=0.26, color='#d00000', linestyle='--', linewidth=2)
# plt.axvline(x=0.13, color='#a2aebb', linestyle='--', linewidth=2)
plt.axvline(x=0.06, color='#ffba08', linestyle='--', linewidth=2)
plt.axvline(x=0.07, color='#3f88c5', linestyle='--', linewidth=2)

# Show the legend
# plt.legend()

# Display the plot
plt.savefig(f'./images/{domain}.png', dpi=600)
plt.show()
```
